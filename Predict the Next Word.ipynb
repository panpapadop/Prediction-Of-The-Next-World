```python
import array as arr
import numpy as np
import random
import string

puncs = set(string.punctuation)

def remove(duplicate): 
    final_list = [] 
    for num in duplicate: 
        if num not in final_list: 
            final_list.append(num) 
    return final_list

uniquewords = []
words=[]


with open('Input.txt', 'r') as f:
    contents = f.readlines()
    for line in contents:
        for c in line.split():
            if c[-1:] in puncs:
                punct = c[-1:] 
                c=c[:-1]
                words.append(c)
                words.append(punct)
            elif c[:1] in puncs:
                punct = c[:1] 
                c=c[1:]
                words.append(punct)
                words.append(c)
                
            else:
                words.append(c)


#Λιστα με τις καταστασεις
uniquewords = remove(words)

print("To sinolo twn katastasewn einai: ", len(uniquewords))
print("Oi katastaseis einai : ",uniquewords)


#Δημιουργια πινακα Markhov
markhov = np.zeros((len(uniquewords),len(uniquewords)))



#Ευρεση πιθανοτητων μεταβασης για καθε στηλη του πινακα Markhov
for i in range(len(uniquewords)):
    counter  = 0
    prob = 0
    nextword = []
    nextindex = []
    for w in range(len(words)-1):
        if uniquewords[i]==(words[w]):
            nextword.append(words[w+1])
            index = uniquewords.index(words[w+1])
            nextindex.append(index)
            counter= counter +1
    for k in range(len(nextindex)):
        prob = nextindex.count(nextindex[k])/counter
        markhov[nextindex[k],i]= prob


print("Ta athrismata ths kathe sthlhs einai:")
print(markhov.sum(axis=0))

#Αρχη της παραγραφου
firstWord = "this"
indexW = uniquewords.index("this")
keimeno=["This"]

#Ευρεση επομενης λεξης της παραγραφου με βαση ενος δειγματος uniform(0,1)
#και της καταλληλης στηλης του πινακα Markhov
for k in range(99):
    uni = np.random.uniform(0,1,len(uniquewords))
    t=0
    for i in range(len(markhov)):
        t = t+markhov[i,indexW]
        if uni[indexW]<=t:
            #Αν η τελευταια λεξη ειναι "." η επομενη να ξεκιναει με κεφαλαιο
            if keimeno[-1]==".":
                nextwordIs = uniquewords[i].title()
                keimeno.append(nextwordIs)
                indexW=i
            else:
                nextwordIs = uniquewords[i]
                keimeno.append(nextwordIs)
                indexW=i
            break


print(keimeno)  

```
